```python
import pandas as pd
df = pd.read_csv('reviews_cronos_groep.csv')
pd.set_option("display.max_rows", None)
df.head(4)
```

```python
from textblob import TextBlob
rating =[]
sentiment = []
rating_user =[]
for ind in df.index:
     text = df['opinion'][ind]
     blob = TextBlob(str(text))
     score = blob.sentiment.polarity# -1 negatief 1 positief
     rating.append(score)
     if score > 0:          
        if score >= 0.5:
            if score <0.6:
                sentiment.append('neutral')
            else:
                sentiment.append('positive')
        else:
            sentiment.append('positive')
     else:
         sentiment.append("negative")


for ind in df.index:
     text = df['rating'][ind]
     
     if  text > 2:
            rating_user.append('positive')
     else:
           rating_user.append('negative')

          

df['sentiment_score'] = sentiment
df['score'] = rating
df['rating_user'] = rating_user
pd.set_option("display.max_rows", None)

pd.set_option("display.max_rows", None)


    
```

```python
df['rating'] = df['rating'].astype(str)
df.to_csv('reviews_cronos_groep2.csv', mode='a', header=False)
```

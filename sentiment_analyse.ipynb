```python
import pandas as pd
df = pd.read_csv('reviews_cronos_groep.csv')
df.head()
```

```python
from textblob import TextBlob
import math
rating =[]
sentiment = []
rating_sentiment =[]
for ind in df.index:
     text = df['opinion'][ind]

     blob = TextBlob(text)
     score = blob.sentiment.polarity# -1 negatief 1 positief
     rating.append(score)
     if score == 0:
         sentiment.append('neutral')
     if score > 0:          
        sentiment.append('positive')
     if score < 0:
        sentiment.append("negative")
       
     
         
for ind in df.index:
     text = df['rating'][ind]
     text1 = df['source'][ind]
     if text >= 3:
           rating_sentiment.append('positive')
     if text1 == "beyondgaming":
           rating_sentiment.append("NaN")
     if text < 3:
           rating_sentiment.append('negative')
         

df['score_sentiment'] = sentiment
df['score'] = rating
df['rating_sentiment'] = rating_sentiment
df.head()
    
```

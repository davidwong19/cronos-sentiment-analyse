```python
import pandas as pd
df = pd.read_csv('headcount.csv')
df.head()
```

```python
df['opinion']
```

```python
from textblob import TextBlob
rating =[]
sentiment = []
for ind in df.index:
     text = df['opinion'][ind]
     blob = TextBlob(text)
     score = blob.sentiment.polarity# -1 negatief 1 positief
     rating.append(score)
     if score > 0:          
        if score >= 0.5:
            if score <0.6:
                sentiment.append('neutral')
            else:
                sentiment.append('positive')
        else:
            sentiment.append('positive')
     else:
         sentiment.append("negative")


df['sentiment'] = sentiment
df['score'] = rating
df.head()
    
```
